## Refresher on Statistical Decision Theory
Formally, the problem we are trying to solve is that of estimating a function $f(X)$ that predicts the values of $Y$ given the input $X$ for a loss function $L(Y, f(X))$.

Various machine learning algorithms make different assumptions about the properties of $f(X)$. For example:

- Linear regression assumes that $f(X)$ is well approximated by a global function, meaning estimations are independent of their local neighborhood.

- K-nearest neighbors assumes that $f(X)$ is well approximated by a local function, meaning estimations are entirely dependent on their local neighborhood.

Global and local functions, and everything inbetween, have their own advantages and disadvantages, which is what we'll look at with the algorithms we learn today.

# Linear Methods
Linear regression for classification runs into issues such as masking. So we can use linear discriminant analysis to improve on this situation. From Bayes Rule we we have that

$$
P(G=k | X = x) = \frac{f_k(x) \pi_k}{\sum_{l=1}^{K}f_l(x) \pi_l}
$$

By assuming that the data originate from normal distributions sharing a covariance matrix, LDA arises (a covariance matrix is a matrix representing the level of dispersion or association between two variables). This gives us the decision function

$$
\delta_k(x) = x^T \Sigma ^{-1}\mu_k - \frac{1}{2}\mu_k^T \Sigma^{-1}\mu_k + log \pi_k
$$

We can extend this to quadratic discriminant analysis.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_moons, make_circles, make_classification
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis, LinearDiscriminantAnalysis
from sklearn.linear_model import LogisticRegression

def plot_classifiers(names, classifiers):
    X, y = make_classification(n_features=2, n_redundant=0, n_informative=2,
                           random_state=1, n_clusters_per_class=1)
    rng = np.random.RandomState(2)
    X += 2 * rng.uniform(size=X.shape)
    linearly_separable = (X, y)

    datasets = [make_moons(noise=0.3, random_state=0),
                make_circles(noise=0.2, factor=0.5, random_state=1),
                linearly_separable
                ]
    h = .02  # step size in the mesh
    figure = plt.figure(figsize=(27, 9))
    i = 1
    # iterate over datasets
    for ds_cnt, ds in enumerate(datasets):
        # preprocess dataset, split into training and test part
        X, y = ds
        X = StandardScaler().fit_transform(X)
        X_train, X_test, y_train, y_test = \
            train_test_split(X, y, test_size=.4, random_state=42)

        x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
        y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
        xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                             np.arange(y_min, y_max, h))

        # just plot the dataset first
        cm = plt.cm.RdBu
        cm_bright = ListedColormap(['#FF0000', '#0000FF'])
        ax = plt.subplot(len(datasets), len(classifiers) + 1, i)
        if ds_cnt == 0:
            ax.set_title("Input data")
        # Plot the training points
        ax.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=cm_bright,
                   edgecolors='k')
        # and testing points
        ax.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap=cm_bright, alpha=0.6,
                   edgecolors='k')
        ax.set_xlim(xx.min(), xx.max())
        ax.set_ylim(yy.min(), yy.max())
        ax.set_xticks(())
        ax.set_yticks(())
        i += 1

        # iterate over classifiers
        for name, clf in zip(names, classifiers):
            ax = plt.subplot(len(datasets), len(classifiers) + 1, i)
            clf.fit(X_train, y_train)
            score = clf.score(X_test, y_test)

            # Plot the decision boundary. For that, we will assign a color to each
            # point in the mesh [x_min, x_max]x[y_min, y_max].
            if hasattr(clf, "decision_function"):
                Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
            else:
                Z = clf.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:, 1]

            # Put the result into a color plot
            Z = Z.reshape(xx.shape)
            ax.contourf(xx, yy, Z, cmap=cm, alpha=.8)

            # Plot also the training points
            ax.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=cm_bright,
                       edgecolors='k')
            # and testing points
            ax.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap=cm_bright,
                       edgecolors='k', alpha=0.6)

            ax.set_xlim(xx.min(), xx.max())
            ax.set_ylim(yy.min(), yy.max())
            ax.set_xticks(())
            ax.set_yticks(())
            if ds_cnt == 0:
                ax.set_title(name)
            ax.text(xx.max() - .3, yy.min() + .3, ('%.2f' % score).lstrip('0'),
                    size=15, horizontalalignment='right')
            i += 1

    plt.tight_layout()
    plt.show()
```

```python
names = ["LDA", "QDA"]


classifiers = [
    LinearDiscriminantAnalysis(),
    QuadraticDiscriminantAnalysis()
]

plot_classifiers(names, classifiers)
```

# Logistic Regression
Another classifier that we can derive from the log odds is the logistic regression classifier, which gives a smooth function of the probability of a value belonging to a class.

$$
f(x) = \frac{e^x}{1 + e^{-x}}
$$

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/8/88/Logistic-curve.svg/320px-Logistic-curve.svg.png">

The derivation of this is simple, as for any class $k$ we have:

$$
\log \frac{Pr(G = k | X = x)}{Pr(G = K | X = x)} = \beta_{k0} + \beta^Tx
$$

Which should be familiar from linear regression with one variable.


Taking the exponential of this gives us the following, which is similar to the logistic function.

$$
Pr(G=k | X = x) = \frac{exp(\beta_{k0} + \beta_k^T x)}{1 + \sum_{l=1}^{K-1 }exp(\beta_{l0} + \beta_l^T x)}
$$

```python
names = ["Logistic Regression"]


classifiers = [
    LogisticRegression()
]

plot_classifiers(names, classifiers)
```

# Problems with this?

You may have been wondering this whole time, how we can be making this strong assumptions about the shape of the data. In both of these models, we assume there is a monotone transformation of the data that is linear, such as the log function(monotone means entirely decreasing or increasing). For LDA, we further assume that the data come from normal distributions and have equal covariance matrices.

Are these valid assumptions?

# No! 

They're probably not valid assumptions about the data, unless we've generated the data to fit these assumptions. 

So how can we justify these assumptions when we can use a model like KNN that doesn't make any such assumptions?

The curse of dimensionality.

In one dimension, to cover 10% of the data, we need to cover 10% of the line. 

In two dimensions, to cover 10% of the data, we need to cover 19% of the area.

In three dimensions, to cover 10% of the data, we need to cover 27% of the volume.

By the time we're up to 10 dimensions, we need to cover 80% of the space just to get 10% of the data.

We can calculate the average distance between data points to be 

$$
   d(p,N) = (1 - \frac{1}{2}^{1/N}) ^ {1/p}
$$

This means that we have a very sparse set of data, and local functions such as KNN become worse and worse the more dimensions we add. 

# Model Assessment and Validation
We need to know how to choose our models and assess whether they're working. First we need to introduce two definitions.

## Model Selection
Estimating the performance of different models in order to choose the best one.

## Model Assessment
Having chosen a final model, estimating its prediction error on new data.

# Dream situation: train - validation - test

In the ideal situation, we have enough data to split the training set into three portions. 
- The first portion (around 50%) is used to train the models.
- The second portion (around 25%) is used to estimate the prediction error for the model (to see if there's an increase in accuracy). This is used for fine tuning the hyper-parameters of the models. 
- The third portion (the remaining 25%) is used to test the generalization error of the model on entirely new data it has not seen.

In a lot of situations we aren't going to have enough data to do this.  This is common when it's expensive to collect data (such as clinical trials) and we don't want to not use half of our data for training. We'll look at cross validation now, but you can also use Bayesian Approaches and the Bootstrap for this.

# Trees

Now going to look at tree based classification methods. 

At the basic level, a tree splits the domain into rectangles with each leaf being a simple model, such as a class. At each branch, a question with multiple outcomes further splits the domain up. (Example on board)

Trees aren't great at fitting complex data for a few reasons. Significantly, they're extermely unstable - any error at the top of the tree is propogated downwards. Also as an aside, when using regression trees, we expect the generating function of the data to be smooth, without jumps - regression trees can't do that.

# Improvements on trees

There are two popular ways of improving on decision trees. Gradient boosting, and random forests, which we will quickly introduce.

## Gradient boosting
In gradient boosting, we grow a number of small trees, or other weak learners, (sometimes even stumps) and take a weighted committee vote from them, ie. the most popular vote from all the trees is the class we allocate to. A weak learner is a model that slightly improves over random guessing. This method takes low variance, high bias models and reduces the bias.

A famous gradient boosting model is adaboost.

## Random forests
Random forests involve fully grown trees that are ensembled with gradient boosting. The difference is that here, we are taking high variance, low bias trees and reducing the bias. 

```python
names = ["Decision Tree", "Random Forest", "AdaBoost"]


classifiers = [
    DecisionTreeClassifier(max_depth=5),
    RandomForestClassifier(max_depth=5, n_estimators=10, max_features=1),
    AdaBoostClassifier(),
]

plot_classifiers(names, classifiers)
```

Very simple challenge today that should let you get an intuitive (if not graphical) idea of how the models work and how to work with them.

Using the MNIST digits dataset, what is the best classification accuracy you can get?

```python
from sklearn.datasets import load_digits

digits = load_digits()

images_and_labels = list(zip(digits.images, digits.target))
for index, (image, label) in enumerate(images_and_labels[:4]):
    plt.subplot(2, 4, index + 1)
    plt.axis('off')
    plt.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
    plt.title('Training: %i' % label)

```

```python
# need to use 1 dimension array of values for now  - we lose spatial information here
data = digits['data'] # 1797 instances of 64 variables corresponding to an 8x8 image
```

```python
# example from the sklearn documentation
from sklearn import svm, metrics
n_samples = len(digits.images)
data = digits.images.reshape((n_samples, -1))

# Create a classifier: a support vector classifier
classifier = svm.SVC(gamma=0.001)

# We learn the digits on the first half of the digits
classifier.fit(data[:n_samples // 2], digits.target[:n_samples // 2])

# Now predict the value of the digit on the second half:
expected = digits.target[n_samples // 2:]
predicted = classifier.predict(data[n_samples // 2:])

print("Classification report for classifier %s:\n%s\n"
      % (classifier, metrics.classification_report(expected, predicted)))
print("Confusion matrix:\n%s" % metrics.confusion_matrix(expected, predicted))

images_and_predictions = list(zip(digits.images[n_samples // 2:], predicted))
for index, (image, prediction) in enumerate(images_and_predictions[:4]):
    plt.subplot(2, 4, index + 5)
    plt.axis('off')
    plt.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
    plt.title('Prediction: %i' % prediction)

plt.show()
```
